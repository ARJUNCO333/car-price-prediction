```python
#converton of string into numerical data ==> get_dummy
dummy=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
dummy
```

```python
# combining the 2 dataframes. 
dfe=pd.concat([df,dummy],axis=1)
dfe
```

```python
df.isna().sum()
```

```python
# in the new dataframe it contain both converted and the original.so we have to remove the already existing ones

dfe.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price'],axis=1,inplace=True)
dfe
```

```python
#string replace
dfe['Mileage']=dfe['Mileage'].str.replace('km/kg','')
dfe['Mileage']=dfe['Mileage'].str.replace('kmpl','')
dfe
```

```python
#string replace
dfe['Engine']=dfe['Engine'].str.replace('CC','')
dfe['Power']=dfe['Power'].str.replace('bhp','')
dfe
```

```python
#changing null values into 0
dfe['Mileage']=dfe['Mileage'].str.replace('null','0')
dfe['Power']=dfe['Power'].str.replace('null','0')
dfe['Engine']=dfe['Engine'].str.replace('null','0')
dfe
```

```python
# datatype checking
dfe.dtypes             
```

```python
#converting object values into integer or float.
dfe['Mileage']=dfe['Mileage'].astype(float)
dfe['Engine']=dfe['Engine'].astype(float)
dfe['Power']=dfe['Power'].astype(float)
dfe.dtypes
```

```python
dfe.isna().sum()
```

```python
#replacing the zero values into
import numpy as np
dfe.loc[dfe.Engine==0,'Engine']=np.NaN    #missing data
dfe.loc[dfe.Power==0,'Power']=np.NaN
dfe.loc[dfe.Mileage==0,'Mileage']=np.NaN
```

```python
dfe.isna().sum()
```

```python
# handling missing values using mean and mode
dfe['Engine']=dfe['Engine'].fillna(dfe["Engine"].mean())
dfe['Mileage']=dfe['Mileage'].fillna(dfe['Mileage'].mean())
dfe['Power']=dfe['Power'].fillna(dfe['Mileage'].mean())
dfe['Seats']=dfe['Seats'].fillna(dfe['Seats'].mode()[0])
dfe.isna().sum()
```

```python
#input and output of training data
x=dfe.drop(['Price'],axis=1)
x
```

```python
y=dfe["Price"]
y
```
